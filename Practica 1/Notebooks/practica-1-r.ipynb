## (a) Create an array of 10 random integers between 1 and 100, and then calculate the mean and standard deviation of the array.

```r
# Generar un array de 10 enteros aleatorios entre 1 y 100
array <- sample(1:100, 10, replace = TRUE) #replace=TRUE permite que un numero generado aleatorio se pueda re[etir]

# Calcular la media y desviación estándar del array
mean_value <- mean(array)
sd_value <- sd(array)

# Imprimir los resultados
cat("Array:", array, "\n")
cat("Mean:", mean_value, "\n")
cat("Standard Deviation:", sd_value, "\n")
```

## (b) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column.


```r
# Crea la matriz de 3 filas y 4 columnas con valores enteros aleatorios entre 1 y 100
matriz <- matrix(sample(1:100, 3*4, replace = TRUE), nrow = 3, ncol = 4)

# Imprime la matriz
print(matriz)

# Calcula la suma de cada fila
sumas_filas <- rowSums(matriz)
cat("Suma de cada fila:", sumas_filas, "\n")

# Calcula la suma de cada columna
sumas_columnas <- colSums(matriz)
cat("Suma de cada columna:", sumas_columnas, "\n")
```

## (c) Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than or equal to 30.

## (d) Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame

```r
# Cargar el archivo CSV en un DataFrame
df <- read.csv("/kaggle/input/clinic/clinic (1).csv", sep=';')

# Calcular la media de una columna
mean_value <- mean(df$age)

# Calcular la mediana de una columna
median_value <- median(df$age)

# Calcular la moda de una columna
mode_value <- names(table(df$age))[table(df$age) == max(table(df$age))]

# Imprimir los resultados
print(paste("Media:", mean_value))
print(paste("Mediana:", median_value))
print(paste("Moda:", mode_value))
```

## (s) Create a new attribute TempC which represents the temperature values in Celsius.

```r
# Crear el nuevo atributo TempC en grados Celsius
df <- df %>% mutate(TempC = (Temp - 32) * (5/9))

# Mostrar el resultado
print(df)
```

## (t) Inspect which were the 30 hottest days

```r
# Ordenar el DataFrame por temperatura en orden descendente
df_sorted <- df %>% arrange(desc(Temp))

# Seleccionar las primeras 30 filas que corresponden a los 30 días más calurosos
hottest_days <- df_sorted %>% slice(1:30)

# Mostrar los 30 días más calurosos
print(hottest_days)
```

### (x) Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone.

```r

df_summary <- df %>%
  group_by(Month) %>%
  summarise(TempC_avg = mean(TempC),
            Wind_avg = mean(Wind),
            SolarR_avg = mean(Solar.R),
            Ozone_avg = mean(Ozone))

df_summary
```
